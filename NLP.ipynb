```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.txt' , sep = ";" , header = None , names=['text','emotion'])
```

```python
unique_emotion =df['emotion'].unique()
emotion_number ={}
i=0
for emotion in unique_emotion:
  emotion_number[emotion]=i
  i+=1
```

```python
df['emotion']= df['emotion'].map(emotion_number)
```

```python
df.head()
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
import string
def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_number(txt):
  new = "" # Initialize with empty string
  for i in txt:
    if not i.isdigit(): # Added parentheses and dot
      new = new + i # Corrected assignment
  return new
df['text'] = df['text'].apply(remove_number) # Removed extra dot
```

```python
def remove_emoji(txt):
  new =""
  for i in txt:
    if i.isascii():
      new+=i
  return new
df['text'] = df['text'].apply(remove_emoji)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
len(stop_words)
```

```python
def remove(txt):
  words = txt.split()
  new =[]
  for i in words:
    if not i in stop_words:
      new.append(i)
  return ''.join(new)
df['text']=df['text'].apply(remove)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
bow_Vectorize = CountVectorizer()
X_train_bow = bow_Vectorize.fit_transform(X_train)
X_test_bow = bow_Vectorize.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train_bow,y_train)
```

```python
pred_nb =nb_model .predict(X_test_bow)
```

```python
pred_nb
```

```python
print(accuracy_score(y_test,pred_nb))
```

```python
Tfidf_Vectorizer =TfidfVectorizer()
X_train_tfidf = Tfidf_Vectorizer.fit_transform(X_train)
X_test_tfidf = Tfidf_Vectorizer.transform(X_test)
```

```python
nb2_model = MultinomialNB()
nb2_model.fit(X_train_tfidf,y_train)
```

```python
y_pred = nb2_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
log_model.fit(X_train_tfidf,y_train)
```

```python
log_pred = log_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,log_pred))
```
